```python
def follow(filename):
    with open(filename) as fd:
        while True:
            line = fd.xreadlines()
            yield next(line).strip()
```

```python
# genmultiplex.py
#
# Multiplex many different generators into a single real-time stream using threads.

import time
import threading, Queue

#from genqueue import *
#from gencat import *

def sendto_queue(source, out_q):
    for s in source:
        out_q.put(s)
    
def genfrom_queue(in_q):
    while True:
        try:
            value = in_q.get(block=True, timeout=2)
        except Queue.Empty:
            break
        else:
            yield value

def gen_cat(sources):
    for source in sources:
        for s in source:
            yield s
    
def multiplex(sources):
    in_q = Queue.Queue()
    consumers = []
    for s in sources:
        thr = threading.Thread(target=sendto_queue,
                               args=(s,in_q))
        thr.start()
        consumers.append(genfrom_queue(in_q))
    return gen_cat(consumers)

if __name__ == '__main__':
    #import follow
    a_log = follow("a.log")
    b_log = follow("b.log")
    for line in multiplex([a_log,b_log]):
        print line
```

```python
# broadcast.py
#
# Broadcast a generator source to a collection of consumers

def broadcast(source, consumers):
    for item in source:
        for c in consumers:
            c.send(item)


# Example
if __name__ == '__main__':

    class Consumer(object):
        def send(self,item):
            print self, "got", item

    c1 = Consumer()
    c2 = Consumer()
    c3 = Consumer()

    #from follow import *
    lines = follow("b.log")
    broadcast(lines,[c1,c2,c3])
```
