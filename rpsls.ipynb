Rock, paper, scissors, lizard, Spock game implemented for Coursera

```python
from math import sqrt
from random import randrange
```

```python
choices = ['rock', 'Spock', 'paper', 'lizard', 'scissors']
choices = ['rock', 'paper']
c_len = len(choices)
margin = int(sqrt(c_len))
```

```python
def rpsls(choice):
    if choice not in choices:
        raise (ValueError, "%s is not a valid choice." % choice)
        
    def _rpsls(pc, cc):
        num = (pc - cc) % c_len
        if num == 0:
            winner = "Player and computer tie!"
        elif num <= margin:
            winner = "Player wins!"
        else:
            winner = "Computer wins!"
        return winner
    
    pc = choices.index(choice)
    cc = randrange(c_len)
    
    #print
    #print "Player chooses %s" % choice
    #print "Computer chooses %s" % choices[cc]
    #print _rpsls(pc, cc)
    #print
    print "\n".join(["%s, %s, %s" % (choices[pc], choices[cc], _rpsls(pc, cc)) \
            for pc in xrange(c_len) for cc in xrange(c_len)])
    
```
